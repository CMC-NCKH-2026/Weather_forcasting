```python
import pandas as pd
import numpy as np
```

# Set up

```python
weather_data = pd.read_csv('../data/weatherHistory.csv')
weather_data.head()
```

```python
weather_data.info()
```

```python
def create_features(df):
    df = df.copy()
    df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
    df['Hour'] = df['Formatted Date'].dt.hour
    df['Day'] = df['Formatted Date'].dt.day
    df['Month'] = df['Formatted Date'].dt.month
    df['Year'] = df['Formatted Date'].dt.year
    df['Weekday'] = df['Formatted Date'].dt.weekday
    return df

processed_weather_data = create_features(weather_data)
processed_weather_data.head()
```

```python
processed_weather_data['Pressure (millibars)'].replace(0, np.nan, inplace=True)
processed_weather_data[processed_weather_data['Pressure (millibars)'].isna()]
```

```python
processed_weather_data['Precip Type'].fillna('none', inplace=True)
processed_weather_data[processed_weather_data['Precip Type'] == 'none']
```

```python
processed_weather_data.query('`Loud Cover` == 0')
```

```python
processed_weather_data.drop('Loud Cover', axis=1, inplace=True)
processed_weather_data
```

# Data Export

```python
processed_weather_data.to_csv('../data/processed_weather_data.csv', index=False)
```
