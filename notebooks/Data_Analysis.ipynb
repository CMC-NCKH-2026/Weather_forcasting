# Analyzing Data through time
### Questions need to be answered
1. The temperature and humidity throughout the years
2. The weather pattern monthly throughout the year?
3. The weather pattern hourly throughout the day?
4. Yearly Variation: How has the average wind speed changed over the years?
5. Monthly Difference: In which month of the year is the largest gap between the Apparent Temperature (째C) and the Temperature (째C)?
6. What is the trend of average Pressure change over the years?
7. How has the average Visibility changed over the years?
8. What is the pattern of average Wind Speed change by hour of the day?
9. What is the pattern of average Visibility by hour of the day (e.g., does fog often appear in the early morning)?
10. What is the hourly pattern of the difference between Apparent Temperature and Temperature throughout the day?
11. What is the pattern of average Pressure by month of the year?
12. Is there any difference in average temperature across the days of the week (Weekday)?
13. How has the frequency of 'Snow' (in the Precip Type column) changed over the years?
14. In which month of the year is the average Visibility the lowest?
15. How does the average daily temperature range (Max Temp - Min Temp) change across the months?

```python
#The temperature and humidity throughout the years
yearly_avg = df[['Temperature (C)', 'Humidity']].resample('Y').mean()
yearly_avg
```

```python
# The weather pattern monthly throughout the year?
monthly_cycle = df.groupby('Month')[['Temperature (C)', 'Humidity']].mean()
monthly_cycle
```

```python
#The weather pattern hourly throughout the day?
hourly_cycle = df.groupby('Hour')[['Temperature (C)', 'Humidity']].mean()
hourly_cycle
```

```python
#Yearly Variation: How has the average wind speed changed over the years?
Yearly_wind_speed = df[['Wind Speed (km/h)']].resample('Y').mean()
Yearly_wind_speed
```

```python
#Monthly Difference: In which month of the year is the largest gap between the Apparent Temperature (째C) and the Temperature (째C)?
monthly_diff= df.groupby('Month')[['Apparent Temperature (C)', 'Temperature (C)']].max()
monthly_diff['Temp_Gap'] = monthly_diff['Apparent Temperature (C)'] - monthly_diff['Temperature (C)']
monthly_diff.groupby("Month")['Temp_Gap']
monthly_diff
```

```python
#What is the trend of average Pressure change over the years?
Yearly_pressure = df[['Pressure (millibars)']].resample('Y').mean()
Yearly_pressure
```

```python
#How has the average Visibility changed over the years?
Yearly_visibility = df[['Visibility (km)']].resample('Y').mean()
Yearly_visibility
```

```python
#What is the pattern of average Wind Speed change by hour of the day?
hourly_wind_speed = df.groupby('Hour')[['Wind Speed (km/h)']].mean()
hourly_wind_speed
```

```python
#What is the pattern of average Visibility by hour of the day (e.g., does fog often appear in the early morning)?
hourly_visibility = df.groupby('Hour')[['Visibility (km)']].mean()
hourly_visibility
```

```python
#What is the hourly pattern of the difference between Apparent Temperature and Temperature throughout the day?
hourly_temp_diff = df.groupby('Hour')[['Apparent Temperature (C)', 'Temperature (C)']].mean()
hourly_temp_diff['Temp_Diff'] = hourly_temp_diff['Apparent Temperature (C)'] - hourly_temp_diff['Temperature (C)']
hourly_temp_diff[['Temp_Diff']]
```

```python
#What is the pattern of average Pressure by month of the year?
monthly_pressure = df.groupby('Month')[['Pressure (millibars)']].mean()
monthly_pressure
```

```python
#Is there any difference in average temperature across the days of the week (Weekday)?
weekday_temp = df.groupby('Weekday')[['Temperature (C)']].mean()
weekday_temp
```

```python
#How has the frequency of 'Snow' (in the Precip Type column) changed over the years?
snow_frequency = df[df['Precip Type'] == 'snow'].resample('Y').size()
snow_frequency
```

```python
#In which month of the year is the average Visibility the lowest?
monthly_visibility = df.groupby('Month')[['Visibility (km)']].mean()
monthly_visibility
```

```python
#How does the average daily temperature range (Max Temp - Min Temp) change across the months?
daily_temp_range = df.resample('D')[['Temperature (C)']].agg(['max', 'min'])
daily_temp_range['Temp_Range'] = daily_temp_range['Temperature (C)']['max'] - daily_temp_range['Temperature (C)']['min']
monthly_temp_range = daily_temp_range[['Temp_Range']].resample('M').mean()
monthly_temp_range
```

# Correlation Analysis
### Question need to be answered
1. How are weather elements related to each other?
2. Correlation between Wind Speed and Visibility
3. Correlation Level of Humidity

```python
numeric_cols = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Visibility (km)', 'Pressure (millibars)']
corr_matrix = df[numeric_cols].corr()
corr_matrix
```

```python
#Correlation between Wind Speed and Visibility
corr_between_wind_visibility= df[['Wind Speed (km/h)','Visibility (km)']].corr()
print("Correlation between Wind Speed and Visibility:")
corr_between_wind_visibility
```

```python
#Correlation Level of Humidity
numeric_cols = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Visibility (km)', 'Pressure (millibars)']
corr_df = df[numeric_cols]
full_corr_matrix =corr_df.corr()
humidity_correlations = full_corr_matrix['Humidity']
print("Correlation Level of Humidity:")
print(humidity_correlations.drop('Humidity'))
```

# Categorical Analysis
### Questions need to be answered
1. Most common summaries
2. Comparing between the average temperature of (Summary) Foggy and Clear
3. Comparing between the average temperature of (Precip Type) rain and snow
4. Comparing between the average humidity of (Precip Type) rain snow and none
5. Comparing between the average wind speed of (sumary) Breezy and Mostly Cloudy, Overcast, Mostly Cloudy, Partly Cloudy

```python
# Most common summaries
df['Summary'].value_counts().head(10)
```

```python
# Comparing between the average temperature of (Summary) Foggy and Clear
temp_by_summary = df.groupby('Summary')['Temperature (C)'].mean().sort_values()
temp_by_summary[['Foggy', 'Clear']]
```

```python
#Comparing between the average temperature of (Precip Type) rain snow and none
temp_by_preciptype=df.groupby('Precip Type')['Temperature (C)'].mean().sort_values()
temp_by_preciptype[['rain', 'snow','none']]
```

```python
#Comparing between the average humidity of (Precip Type) rain snow and none
humidity_by_preciptype=df.groupby('Precip Type')['Humidity'].mean().sort_values()
humidity_by_preciptype[['rain', 'snow','none']]
```

```python
#Comparing between the average wind speed of (sumary) Breezy and Mostly Cloudy, Overcast, Mostly Cloudy, Partly Cloudy
wind_by_summary = df.groupby('Summary')['Wind Speed (km/h)'].mean().sort_values()
wind_by_summary[['Breezy', 'Mostly Cloudy', 'Overcast', 'Partly Cloudy','Breezy and Mostly Cloudy']]
```
