# Set up

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df.set_index('Formatted Date', inplace=True)
```

# Analyzing Data through time
### Questions need to be answered
1. The temperature and humidity throughout the years
2. The weather pattern monthly throughout the year?
3. The weather pattern hourly throughout the day?
4. Yearly Variation: How has the average wind speed changed over the years?
5. Monthly Difference: In which month of the year is the largest gap between the Apparent Temperature (째C) and the Temperature (째C)?


```python
#The temperature and humidity throughout the years
yearly_avg = df.groupby('Year')[['Temperature (C)', 'Humidity']].mean()
yearly_avg
```

```python
# The weather pattern monthly throughout the year?
monthly_cycle = df.groupby('Month')[['Temperature (C)', 'Humidity']].mean()
monthly_cycle
```

```python
#The weather pattern hourly throughout the day?
hourly_cycle = df.groupby('Hour')[['Temperature (C)', 'Humidity']].mean()
hourly_cycle
```

```python
#Yearly Variation: How has the average wind speed changed over the years?
Yearly_wind_speed = df[['Wind Speed (km/h)']].resample('Y').mean()
Yearly_wind_speed
```

```python
#Monthly Difference: In which month of the year is the largest gap between the Apparent Temperature (째C) and the Temperature (째C)?
monthly_diff= df.groupby('Month')[['Apparent Temperature (C)', 'Temperature (C)']].max()
monthly_diff['Temp_Gap'] = monthly_diff['Apparent Temperature (C)'] - monthly_diff['Temperature (C)']
monthly_diff.groupby("Month")['Temp_Gap']
monthly_diff
```

# Correlation Analysis
### Question need to be answered
1. How are weather elements related to each other?
2. Correlation between Wind Speed and Visibility
3. Correlation Level of Humidity

```python
numeric_cols = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Visibility (km)', 'Pressure (millibars)']
corr_matrix = df[numeric_cols].corr()
corr_matrix
```

```python
#Correlation between Wind Speed and Visibility
corr_between_wind_visibility= df[['Wind Speed (km/h)','Visibility (km)']].corr()
print("Correlation between Wind Speed and Visibility:")
corr_between_wind_visibility
```

```python
#Correlation Level of Humidity
numeric_cols = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Visibility (km)', 'Pressure (millibars)']
corr_df = df[numeric_cols]
full_corr_matrix =corr_df.corr()
humidity_correlations = full_corr_matrix['Humidity']
print("Correlation Level of Humidity:")
print(humidity_correlations.drop('Humidity'))
```

# Categorical Analysis
### Questions need to be answered
1. Most common summaries
2. Comparing between the average temperature of (Summary) Foggy and Clear
3. Comparing between the average temperature of (Precip Type) rain and snow
4. Comparing between the average humidity of (Precip Type) rain snow and none
5. Comparing between the average wind speed of (sumary) Breezy and Mostly Cloudy, Overcast, Mostly Cloudy, Partly Cloudy

```python
# Most common summaries
summary_common = df['Summary'].value_counts()
summary_common
```

```python
# Comparing between the average temperature of (Summary) Foggy and Clear theo th찼ng
foggy = df[df['Summary'] == 'Foggy']
foggy_tem_avg = foggy.groupby('Month')[['Temperature (C)']].mean()

clear = df[df['Summary'] == 'Clear']
clear_tem_avg = clear.groupby('Month')[['Temperature (C)']].mean()

clear_tem_avg

```

```python
#Comparing between the average temperature of (Precip Type) rain snow and none
rain = df[df["Precip Type"] == 'rain']
rain_tem_avg = rain.groupby('Month')[['Temperature (C)']].mean()

snow = df[df['Precip Type'] == 'snow']
snow_tem_avg = snow.groupby('Month')[['Temperature (C)']].mean()

none_precip = df[df['Precip Type'] == 'none']
none_tem_avg = none_precip.groupby('Month')[['Temperature (C)']].mean()

rain_tem_avg
```

```python
#Comparing between the average humidity of (Precip Type) rain snow and none

rain_humid = df[df["Precip Type"] == 'rain']
rain_humid_avg = rain_humid.groupby('Month')[['Humidity']].mean()

snow_humid = df[df['Precip Type'] == 'snow']
snow_humid_avg = snow_humid.groupby('Month')[['Humidity']].mean()

none_precip_humid = df[df['Precip Type'] == 'none']
none_humid_avg = none_precip_humid.groupby('Month')[['Humidity']].mean()

none_humid_avg
```

```python
#Comparing between the average wind speed of (sumary) Breezy and Mostly Cloudy, Overcast Partly Cloudy
wind_by_summary = df.groupby('Summary')['Wind Speed (km/h)'].mean().sort_values()
selected = ['Breezy', 'Mostly Cloudy', 'Overcast', 'Partly Cloudy']
wind_selected = wind_by_summary[wind_by_summary.index.isin(selected)]
wind_selected
```
