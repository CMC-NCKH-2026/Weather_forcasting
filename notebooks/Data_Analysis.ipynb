# Set up

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df.set_index('Formatted Date', inplace=True)
```

# Analyzing Data through time
### Questions need to be answered
1. The temperature and humidity throughout the years
2. The weather pattern monthly throughout the year?
3. The weather pattern hourly throughout the day?

```python
#The temperature and humidity throughout the years
yearly_avg = df[['Temperature (C)', 'Humidity']].resample('Y').mean()
yearly_avg
```

```python
# The weather pattern monthly throughout the year?
monthly_cycle = df.groupby('Month')[['Temperature (C)', 'Humidity']].mean()
monthly_cycle
```

# Correlation Analysis
### Question need to be answered
How are weather elements related to each other?

```python
numeric_cols = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Visibility (km)', 'Pressure (millibars)']
corr_matrix = df[numeric_cols].corr()
corr_matrix
```

# Categorical Analysis
### Questions need to be answered
1. Most common summaries
2. Comparing between the average temperature of (Summary) Foggy and Clear
3. Comparing between the average temperature of (Precip Type) rain and snow

```python
# Most common summaries
df['Summary'].value_counts().head(10)
```

```python
# Comparing between the average temperature of (Summary) Foggy and Clear
temp_by_summary = df.groupby('Summary')['Temperature (C)'].mean().sort_values()
temp_by_summary[['Foggy', 'Clear']]
```
